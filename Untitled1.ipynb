# Nykaa Marketing Campaign Performance Analysis & ROI Optimization

# Problem Statement

Nykaa runs multiple digital marketing campaigns across different channels such as Instagram, Facebook, Google Ads, Email, and other platforms. The company generates large amounts of campaign data including impressions, clicks, leads, conversions, revenue, and acquisition costs.

The business wants to understand:

Which marketing channels generate the highest conversions?

Which campaigns provide the best return on investment (ROI)?

Which customer segments respond better to campaigns?

Where are users dropping in the marketing funnel?

How can marketing spending be optimized to improve revenue?

The objective of this project is to clean, analyze, and visualize marketing campaign data to identify high-performing campaigns and provide data-driven recommendations for improving marketing effectiveness.

# Project Description

This project analyzes Nykaa’s marketing campaign performance data using Python, SQL, and Power BI.

The analysis focuses on evaluating campaign effectiveness by calculating important marketing KPIs such as:

Click Through Rate (CTR)

Conversion Rate (CVR)

Customer Acquisition Cost (CAC)

Return on Ad Spend (ROAS)

Revenue Performance

Campaign ROI

The project includes data cleaning, exploratory data analysis, statistical analysis, customer segmentation, and dashboard creation to help business stakeholders make better marketing decisions.

```python
for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_rows = df[
        (df[col] < lower_bound) |
        (df[col] > upper_bound)
    ]

    print(f"\nColumn: {col}")
    print(f"Outlier Rows: {len(outlier_rows)}")

    if not outlier_rows.empty:
        display(outlier_rows)
```

```python
num_cols = [
    'Impressions',
    'Clicks',
    'Leads',
    'Conversions',
    'Revenue',
    'Acquisition_Cost',
    'ROI',
    'Engagement_Score'
]

outlier_stats = []

for col in num_cols:
    Q1 = df[col].quantile(0.25)
    Q3 = df[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR

    outlier_count = ((df[col] < lower_bound) | (df[col] > upper_bound)).sum()
    outlier_percent = (outlier_count / len(df)) * 100

    outlier_stats.append([
        col,
        outlier_count,
        round(outlier_percent, 2)
    ])

outlier_df = pd.DataFrame(
    outlier_stats,
    columns=['Column', 'Outlier Count', 'Outlier %']
)

print(outlier_df)
```

```python

num_cols = [
    'Impressions',
    'Clicks',
    'Leads',
    'Conversions',
    'Revenue',
    'Acquisition_Cost',
    'ROI',
    'Engagement_Score'
]


df_clean = df.copy()

mask = pd.Series(True, index=df_clean.index)

for col in num_cols:
    Q1 = df_clean[col].quantile(0.25)
    Q3 = df_clean[col].quantile(0.75)
    IQR = Q3 - Q1

    lower_bound = Q1 - (1.5 * IQR)
    upper_bound = Q3 + (1.5 * IQR)

    mask &= (
        (df_clean[col] >= lower_bound) &
        (df_clean[col] <= upper_bound)
    )


df_clean = df_clean[mask]



print("AFTER OUTLIER REMOVAL")

print("Shape:", df_clean.shape)

rows_removed = len(df) - len(df_clean)
percent_removed = (rows_removed / len(df)) * 100

print("Rows Removed:", rows_removed)
print(f"Percentage Removed: {percent_removed:.2f}%")


for col in num_cols:
    plt.figure(figsize=(6,3))
    plt.boxplot(df_clean[col])
    plt.title(f'{col} - After Outlier Removal')
    plt.grid(True, linestyle='--', alpha=0.5)
    plt.show()


df_clean.to_csv("marketing_campaign_no_outliers.cs", index=False)

print("\nCleaned dataset saved as:")
print("marketing_campaign_no_outliers.csv")
```

```python
# Create a copy
df_channels = df_clean.copy()

# Split channels into lists
df_channels['Channel_Used'] = df_channels['Channel_Used'].str.split(',')

# Convert each channel into a separate row
df_channels = df_channels.explode('Channel_Used')

# Remove extra spaces
df_channels['Channel_Used'] = df_channels['Channel_Used'].str.strip()

# Total conversions by channel
channel_conversions = (
    df_channels.groupby('Channel_Used')['Conversions']
    .sum()
    .sort_values(ascending=False)
)

print(channel_conversions)
```

```python



impressions = df_clean['Impressions'].sum()
clicks = df_clean['Clicks'].sum()
leads = df_clean['Leads'].sum()
conversions = df_clean['Conversions'].sum()


funnel = pd.DataFrame({
    'Stage': ['Impressions', 'Clicks', 'Leads', 'Conversions'],
    'Count': [impressions, clicks, leads, conversions]
})


funnel['Drop_From_Previous'] = funnel['Count'].shift(1) - funnel['Count']
funnel['Drop_%'] = (
    funnel['Drop_From_Previous'] /
    funnel['Count'].shift(1) * 100
).round(2)

print("Marketing Funnel")
print(funnel)


plt.figure(figsize=(8,5))
bars = plt.bar(funnel['Stage'], funnel['Count'])

plt.title('Marketing Funnel Analysis')
plt.xlabel('Funnel Stage')
plt.ylabel('Total Users')


for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f'{int(bar.get_height()):,}',
        ha='center'
    )

plt.tight_layout()
plt.show()


print("\nDrop-off Analysis")
for i in range(1, len(funnel)):
    print(
        f"{funnel.loc[i-1,'Stage']} → {funnel.loc[i,'Stage']}: "
        f"{funnel.loc[i,'Drop_%']}% drop"
    )
```

## Marketing Funnel Performance Analysis

###  Strong Audience Engagement
Out of all the people who saw the ads, **8.05% clicked on them**. Considering the campaign reached more than **2.4 billion impressions**, this is an excellent Click-Through Rate (CTR). It indicates that the advertisements were relevant, engaging, and effective at capturing audience attention.

###  Effective Lead Generation
Among the users who clicked on the ads, **38.14% became leads**. This means that more than one out of every three visitors showed genuine interest in the product or service. The result suggests that the landing pages and campaign offers successfully encouraged users to take the next step.

###  Excellent Conversion Performance
Of all the leads generated, **53.13% ultimately converted into customers or completed the desired action**. Converting more than half of the leads is a strong outcome and indicates an efficient sales process, effective follow-up strategies, or highly qualified leads.

###  Strong Overall Funnel Efficiency
From the initial audience exposure to the final conversion, **1.63% of all impressions resulted in successful conversions**. Given the massive scale of the campaign, this represents strong overall performance and demonstrates that the marketing funnel is effective at every stage.

###  Key Takeaway
The campaign performed strongly throughout the entire customer journey. The ads successfully attracted attention, a significant proportion of visitors became leads, and more than half of those leads converted. Overall, the results indicate a well-optimized marketing strategy with strong engagement, lead generation, and conversion performance.

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a copy
df_channels = df_clean.copy()

# Split channels into separate rows
df_channels['Channel_Used'] = df_channels['Channel_Used'].str.split(',')

# Explode channels
df_channels = df_channels.explode('Channel_Used')

# Remove extra spaces
df_channels['Channel_Used'] = df_channels['Channel_Used'].str.strip()

# Revenue generated per ₹1 spent
df_channels['Revenue_per_Rupee'] = (
    df_channels['Revenue'] /
    df_channels['Acquisition_Cost']
)

# -----------------------------
# TOTAL REVENUE BY CHANNEL
# -----------------------------
channel_revenue = (
    df_channels.groupby('Channel_Used')['Revenue']
    .sum()
    .sort_values(ascending=False)
)

print("Total Revenue by Channel")
print(channel_revenue)

# Plot Total Revenue
plt.figure(figsize=(8,5))
bars = plt.bar(channel_revenue.index, channel_revenue.values)

plt.title('Total Revenue by Marketing Channel')
plt.xlabel('Channel')
plt.ylabel('Total Revenue')

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f'{bar.get_height():,.0f}',
        ha='center',
        va='bottom'
    )

plt.tight_layout()
plt.show()

# -----------------------------
# REVENUE PER ₹1 SPENT
# -----------------------------
channel_efficiency = (
    df_channels.groupby('Channel_Used')['Revenue_per_Rupee']
    .mean()
    .sort_values(ascending=False)
)

print("\nRevenue Generated per ₹1 Spent")
print(channel_efficiency)

# Plot Revenue Efficiency
plt.figure(figsize=(8,5))
bars = plt.bar(channel_efficiency.index, channel_efficiency.values)

plt.title('Revenue Generated per ₹1 Spent by Channel')
plt.xlabel('Channel')
plt.ylabel('Revenue per ₹1 Spent')

for bar in bars:
    plt.text(
        bar.get_x() + bar.get_width()/2,
        bar.get_height(),
        f'{bar.get_height():.2f}',
        ha='center',
        va='bottom'
    )

plt.tight_layout()
plt.show()

# -----------------------------
# TOP & BOTTOM CHANNELS
# -----------------------------
print("\nTop 3 Channels by Revenue")
print(channel_revenue.head(3))

print("\nBottom 3 Channels by Revenue")
print(channel_revenue.tail(3))

print("\nTop 3 Channels by Revenue per ₹1 Spent")
print(channel_efficiency.head(3))

print("\nBottom 3 Channels by Revenue per ₹1 Spent")
print(channel_efficiency.tail(3))
```
